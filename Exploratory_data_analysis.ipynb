# summary of Accidents causes
Majority of the accideents occured on single carriageways, without any physical crossing facilities within 50 meters and in day light,
when there were no high winds and the roads were dry. ***So most natural factors weren't the causes of majority of this accidents but instead we can say that the single carriage ways had a huge part to play***

```python
sns.catplot(x='Accident_Severity', data=Accident, kind='count')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.catplot(x='Day_of_Week',y='Number_of_Vehicles', data=Accident, kind='point')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.catplot(x='Day_of_Week',y='Number_of_Casualties', data=Accident, kind='point')
plt.xticks(rotation=90)
plt.show()
```

# day of the week
Though we see that fewer vehicles were involved in clashes during the weekend, ***the accidents that occurerd during the weekend most especially sunday and Saturday had a lot more casualities***

```python
sns.relplot(x='Date',y='Number_of_Casualties', data=Accident, kind='scatter')
plt.show()
```

```python
sns.relplot(x='Date',y='Number_of_Vehicles', data=Accident, kind='scatter')
plt.show()
```

# changes with time
There has thankfully not been any upward trends over time in the number of casualities or number of vehicles

```python
sns.relplot(x='Time',y='Number_of_Vehicles', data=Accident, kind='scatter')
plt.show()
```
